```python
import requests
from bs4 import BeautifulSoup
import os
import time

os.makedirs("pokemon_sprites", exist_ok=True)

resp = requests.get("https://pokemondb.net/sprites")
soup = BeautifulSoup(resp.text, "html.parser")

all_names = [
    a["href"].split("/")[-1]
    for a in soup.select("a[href^='/sprites/']")
    if a["href"] != "/sprites"
]

all_names = list(dict.fromkeys(all_names))

downloaded = 0
skipped = 0

for name in sorted(all_names):
    if os.path.exists(f"pokemon_sprites/{name}.png"):
        continue
    url = f"https://img.pokemondb.net/sprites/home/normal/{name}.png"
    resp = requests.get(url)
    if resp.status_code == 200:
        with open(f"pokemon_sprites/{name}.png", "wb") as f:
            f.write(resp.content)
        print(f"Downloaded: {name}")
        downloaded += 1
    else:
        print(f"Not found: {name} (status {resp.status_code})")
        skipped += 1
    time.sleep(0.1)

print(f"\nDone. {downloaded} downloaded, {skipped} skipped.")
```

```python
base_url = "https://www.smogon.com/stats/"
months = [
    "2021-01/",
    "2021-02/",
    "2021-03/",
    "2021-04/",
    "2021-05/",
    "2021-06/",
]

target_format = "gen8ou"
target_rating = "1825"
save_dir = "gen8ou-1825.json"

os.makedirs(save_dir, exist_ok=True)

session = requests.Session()
session.headers.update({
    "User-Agent": "Mozilla/5.0"
})

for month in months:
    month_url = urljoin(base_url, month) + "chaos/"
    print(f"Checking {month_url}")

    try:
        response = session.get(month_url, timeout=20)
        response.raise_for_status()
    except requests.RequestException as e:
        print(f"Failed to access {month_url}: {e}")
        continue

    soup = BeautifulSoup(response.text, "html.parser")
    links = [a.get("href") for a in soup.find_all("a", href=True)]

    matching_files = []
    for href in links:
        filename = href.split("/")[-1]
        if target_format in filename and f"-{target_rating}" in filename:
            matching_files.append(href)

    if not matching_files:
        print(f"No matching chaos file found for {month}")
        continue

    for href in matching_files:
        file_url = urljoin(month_url, href)
        filename = os.path.basename(href)
        output_path = os.path.join(save_dir, f"{month.strip('/')}_{filename}")

        try:
            file_response = session.get(file_url, timeout=30)
            file_response.raise_for_status()
        except requests.RequestException as e:
            print(f"Failed to download {file_url}: {e}")
            continue

        with open(output_path, "wb") as f:
            f.write(file_response.content)

        print(f"Saved: {output_path}")
```

```python
import json
import pandas as pd

input_dir = "gen8ou-1825.json"
output_dir = "gen8ou-1825_csv"

os.makedirs(output_dir, exist_ok=True)

for filename in os.listdir(input_dir):
    if not filename.endswith(".json"):
        continue

    input_path = os.path.join(input_dir, filename)

    with open(input_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    rows = []
    info = data.get("info", {})
    pokemon_data = data.get("data", {})

    for pokemon, stats in pokemon_data.items():
        row = {
            "pokemon": pokemon,
            "cutoff": info.get("cutoff"),
            "cutoff_deviation": info.get("cutoff deviation"),
            "number_of_battles": info.get("number of battles"),
        }

        for key, value in stats.items():
            if isinstance(value, dict):
                for subkey, subvalue in value.items():
                    row[f"{key}_{subkey}"] = subvalue
            else:
                row[key] = value

        rows.append(row)

    df = pd.DataFrame(rows)

    output_filename = filename.replace(".json", ".csv")
    output_path = os.path.join(output_dir, output_filename)
    df.to_csv(output_path, index=False)

    print(f"Converted {filename} -> {output_filename}")
```
