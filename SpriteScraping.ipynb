```python
import requests
from bs4 import BeautifulSoup
import os
import time

os.makedirs("pokemon_sprites", exist_ok=True)

resp = requests.get("https://pokemondb.net/sprites")
soup = BeautifulSoup(resp.text, "html.parser")

all_names = [
    a["href"].split("/")[-1]
    for a in soup.select("a[href^='/sprites/']")
    if a["href"] != "/sprites"
]

all_names = list(dict.fromkeys(all_names))

downloaded = 0
skipped = 0

for name in sorted(all_names):
    if os.path.exists(f"pokemon_sprites/{name}.png"):
        continue
    url = f"https://img.pokemondb.net/sprites/home/normal/{name}.png"
    resp = requests.get(url)
    if resp.status_code == 200:
        with open(f"pokemon_sprites/{name}.png", "wb") as f:
            f.write(resp.content)
        print(f"Downloaded: {name}")
        downloaded += 1
    else:
        print(f"Not found: {name} (status {resp.status_code})")
        skipped += 1
    time.sleep(0.1)

print(f"\nDone. {downloaded} downloaded, {skipped} skipped.")
```
