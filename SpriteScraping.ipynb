```python
import requests
from bs4 import BeautifulSoup
import os
import time

os.makedirs("sprites_gen8", exist_ok=True)

# grab full pokemon name list from pokemondb
resp = requests.get("https://pokemondb.net/sprites")
soup = BeautifulSoup(resp.text, "html.parser")
all_names = list(set([
    a["href"].split("/")[-1]
    for a in soup.select("a[href^='/sprites/']")
    if a["href"] != "/sprites"
]))

downloaded = 0
skipped = 0

for name in sorted(all_names):
    url = f"https://img.pokemondb.net/sprites/sword-shield/normal/{name}.png"
    resp = requests.get(url)
    if resp.status_code == 200:
        with open(f"sprites_gen8/{name}.png", "wb") as f:
            f.write(resp.content)
        print(f"Downloaded: {name}")
        downloaded += 1
    else:
        skipped += 1
    time.sleep(0.1)

print(f"\nDone. {downloaded} downloaded, {skipped} skipped (not in Gen 8).")
```
