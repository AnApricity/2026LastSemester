```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
```

```python
input_dir = "gen8ou-1825_csv"

USAGE_DIR = "gen8ou_usage"
ITEMS_DIR = "gen8ou_items"
TEAM_DIR = "gen8ou_teams"
TEAMMATES_DIR = "gen8ou_teammates"

os.makedirs(USAGE_DIR, exist_ok=True)
os.makedirs(ITEMS_DIR, exist_ok=True)
os.makedirs(TEAM_DIR, exist_ok=True)
os.makedirs(TEAMMATES_DIR, exist_ok=True)

files = sorted(glob.glob(os.path.join(input_dir, "*.csv")))

for f in files:
    filename = os.path.basename(f)

    if "suspecttest" in filename.lower():
        continue

    df = pd.read_csv(f)
    cols = df.columns.tolist()

    base_cols = [c for c in ["pokemon", "number_of_battles", "total_battles", "Raw count", "raw_usage"] if c in cols]

    usage_cols = base_cols.copy()
    item_cols = [c for c in base_cols]
    team_cols = [c for c in base_cols]
    teammate_cols = [c for c in base_cols]

    for c in cols:
        c_lower = c.lower()

        if c not in usage_cols and c_lower in ["usage", "weighted usage", "viability ceiling"]:
            usage_cols.append(c)

        if c not in item_cols and "item" in c_lower:
            item_cols.append(c)

        if c not in team_cols and ("team" in c_lower or "teammates" in c_lower):
            team_cols.append(c)

        if c not in teammate_cols and "teammate" in c_lower:
            teammate_cols.append(c)

    if not usage_cols:
        usage_cols = [c for c in cols if c in ["pokemon", "number_of_battles", "total_battles", "Raw count", "raw_usage"]]

    usage_df = df[usage_cols].copy()
    items_df = df[item_cols].copy()
    team_df = df[team_cols].copy()
    teammates_df = df[teammate_cols].copy()

    usage_out = os.path.join(USAGE_DIR, filename.replace(".csv", "_usage.csv"))
    items_out = os.path.join(ITEMS_DIR, filename.replace(".csv", "_items.csv"))
    team_out = os.path.join(TEAM_DIR, filename.replace(".csv", "_teams.csv"))
    teammates_out = os.path.join(TEAMMATES_DIR, filename.replace(".csv", "_teammates.csv"))

    usage_df.to_csv(usage_out, index=False)
    items_df.to_csv(items_out, index=False)
    team_df.to_csv(team_out, index=False)
    teammates_df.to_csv(teammates_out, index=False)

    print(f"Saved:\n  {usage_out}\n  {items_out}\n  {team_out}\n  {teammates_out}\n")
```

```python
usage_files = sorted(glob.glob(os.path.join("gen8ou_usage", "*_usage.csv")))

monthly = []

for f in usage_files:
    usage_df = pd.read_csv(f)
    month = os.path.basename(f).replace("_gen8ou-1825_usage.csv", "").replace("_usage.csv", "")

    if "total_battles" in usage_df.columns:
        total_battles = usage_df["total_battles"].iloc[0]
    else:
        total_battles = usage_df["number_of_battles"].iloc[0]

    if "raw_usage" in usage_df.columns:
        usage_df["raw_percent"] = (usage_df["raw_usage"] / total_battles) * 100
    else:
        usage_df["raw_percent"] = (usage_df["Raw count"] / total_battles) * 100

    usage_df = usage_df[["pokemon", "raw_percent"]].copy()
    usage_df["month"] = month
    monthly.append(usage_df)

all_usage = pd.concat(monthly, ignore_index=True)

summary = (
    all_usage.groupby("pokemon")
    .agg(
        mean_percent=("raw_percent", "mean"),
        sd_percent=("raw_percent", "std")
    )
    .reset_index()
)

june_df = all_usage[all_usage["month"].str.contains("2021-06")][["pokemon", "raw_percent"]].copy()
june_df = june_df.rename(columns={"raw_percent": "june_percent"})

plot_df = summary.merge(june_df, on="pokemon", how="inner")
```

```python
thresholds = np.arange(0, plot_df["june_percent"].max() + 1, 1)

counts_june = np.array([(plot_df["june_percent"] >= t).sum() for t in thresholds])

counts_low = np.array([((plot_df["mean_percent"] - plot_df["sd_percent"]).clip(lower=0) >= t).sum() for t in thresholds])
counts_high = np.array([((plot_df["mean_percent"] + plot_df["sd_percent"]) >= t).sum() for t in thresholds])

plt.figure(figsize=(10, 6))
plt.fill_between(
    thresholds,
    counts_low,
    counts_high,
    color="purple",
    alpha=0.25,
    label="Uncertainty band (mean ± SD)"
)
plt.plot(
    thresholds,
    counts_june,
    color="black",
    linewidth=2,
    label="June 2021"
)

plt.xlabel("Usage Threshold (%)")
plt.ylabel("Number of Pokémon with usage ≥ threshold")
plt.title("Number of Pokémon Above Usage Threshold (Gen 8 OU)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
filtered = plot_df[plot_df["june_percent"] >= 5].sort_values("june_percent", ascending=True)

fig, ax = plt.subplots(figsize=(10, 14))
y = np.arange(len(filtered))

ax.barh(y, filtered["june_percent"], color="green", alpha=1.0, label="June 2021 usage", zorder=2)
ax.errorbar(
    filtered["mean_percent"], y,
    xerr=filtered["sd_percent"],
    fmt="none", ecolor="black", elinewidth=2.5,
    capsize=5, capthick=2.5, zorder=3,
    label="Mean ± SD (January to June 2021)"
)
ax.scatter(filtered["mean_percent"], y, color="white", edgecolors="black", s=40, zorder=4, label="January - June mean")

ax.set_yticks(y)
ax.set_yticklabels(filtered["pokemon"], fontsize=8)
ax.set_xlabel("Usage Percentage (%)", fontsize=12)
ax.set_ylabel("Pokémon", fontsize=12)
ax.set_title(
    "Pokémon Usage in Gen 8 OU — June 2021 (≥5% cutoff)\n"
    "Error bars = uncertainty across all months  |  Dot = all-month mean",
    fontsize=13
)
ax.legend(fontsize=10)
ax.grid(axis="x", linestyle="--", alpha=0.35)
plt.tight_layout()
plt.savefig("gen8ou_june2021_bar_uncertainty.png", dpi=150)
plt.show()
```
