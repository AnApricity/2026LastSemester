```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
os.chdir("2026LastSemester")
```

```python
Stranger = pd.read_csv("Strange.csv")
GTrends = pd.read_csv("Trends.csv")
```

```python
Stranger.columns = Stranger.columns.str.strip()
ratings = Stranger.iloc[:, 1:].values.flatten()
rounded_ratings = pd.Series(ratings).round(1).apply(lambda x: round(x * 2 )/ 2)
rating_counts = rounded_ratings.value_counts().sort_index()
plt.bar(rating_counts.index.astype(str), rating_counts.values, color = "pink")
plt.title("Episode Ratings Bar Graph", color = "red")
plt.xlabel("Rating (Rounded to 0.5)", color = "green")
plt.ylabel("Number of Episodes", color = "turquoise")
plt.show()
```

```python
python = GTrends.iloc[1, 1:].astype(float)
lawyers = GTrends.iloc[0,1:].astype(float)
plt.scatter(python, lawyers, color = "green")
plt.title("Python vs Lawyers google searches Scatterplot", color = "yellow")
plt.xlabel("Number of python searches", color = "aquamarine")
plt.ylabel("Number of lawyer searches", color = "maroon")
plt.show()
```
